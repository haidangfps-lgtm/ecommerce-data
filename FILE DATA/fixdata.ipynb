```python
import pandas as pd
```

```python
# 1. Đọc file dữ liệu ban đầu
df = pd.read_csv('marketplace_category_dayofweek.csv')

# 2. Tạo một Dictionary (Từ điển) để ánh xạ thứ tự các ngày trong tuần
day_mapping = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}

# 3. Sử dụng hàm map() để tạo cột mới 'day_index' dựa trên giá trị của cột 'day_of_week'
df['day_index'] = df['day_of_week'].map(day_mapping)

# 4. Kiểm tra xem dữ liệu đã được thêm đúng chưa
print(df[['day_of_week', 'day_index']].head(10))

# 5. Lưu kết quả ra một file CSV mới để đưa vào phần mềm BI hoặc Dashboard
df.to_csv('marketplace_category_dayofweek_processed.csv', index=False)
```
